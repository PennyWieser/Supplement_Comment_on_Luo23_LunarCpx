```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import Thermobar as pt
```

```python
df_pMELTS=pd.read_excel('NeaveExps.xlsx')
df_pMELTS=df_pMELTS.loc[df_pMELTS['SiO2_MELTS_Cpx']>0]
#df_pMELTS=df_pMELTS.replace(0, np.nan, inplace=True)
```

```python
moon_Exps=pd.read_excel('Cpx-Data.xlsx', sheet_name='Experiments')
moon_comps=pt.calculate_clinopyroxene_components(cpx_comps=moon_Exps)
moon_exp_cpx=moon_Exps.loc[moon_comps['Ca_CaMgFe'].between(0.2, 0.5)]
moon_exp_pig=moon_Exps.loc[~(moon_comps['Ca_CaMgFe'].between(0.2, 0.5))]
newexps=moon_Exps
new_exps_E=newexps.loc[newexps['Author(year)'].str.contains('Elardo')]
new_exps_D=newexps.loc[~(newexps['Author(year)'].str.contains('Elardo'))]
#moon_comps.columns
```

```python
fig, (ax0, ax1, ax2)  = plt.subplots(1, 3, figsize=(12,4))
ax0.plot(df_pMELTS['Al2O3_MELTS_Cpx'], df_pMELTS['Al2O3_Cpx'], '.k', label='Neave Exps')
ax0.plot(new_exps_E['Al2O3_MELTS_Cpx'], new_exps_E['Al2O3_Cpx'], 'dk', mfc='red', label='Lunar Exps Cpx')
ax0.plot(new_exps_D['Al2O3_MELTS_Cpx'], new_exps_D['Al2O3_Cpx'], '^k', mfc='salmon', label='Lunar Exps Pig')

ax0.set_xlabel('Cpx Al$_2$O$_3$ MELTS predicted')
ax0.set_ylabel('Cpx Al$_2$O$_3$ measured')
ax0.plot([2, 13], [2, 13], '-k', label='1:1')


#ax1.plot(df_pMELTS['Na2O_MELTS_Cpx'], df_pMELTS['Na2O_Cpx'], '.k', label='Neave Exps')
ax1.plot(new_exps_E['Na2O_MELTS_Cpx'], new_exps_E['Na2O_Cpx'], 'dk', mfc='red', label='Lunar Exps Cpx')
ax1.plot(new_exps_D['Na2O_MELTS_Cpx'], new_exps_D['Na2O_Cpx'], '^k', mfc='salmon', label='Lunar Exps Pig')

ax2.plot(df_pMELTS['Na2O_MELTS_Cpx'], df_pMELTS['Na2O_Cpx'], '.k', label='Neave Exps')
ax1.set_xlabel('Cpx Na$_2$O MELTS predicted')
ax1.set_ylabel('Cpx Na$_2$O measured')

ax2.set_xlabel('Cpx Na$_2$O MELTS predicted')
ax2.set_ylabel('Cpx Na$_2$O measured')

ax1.plot([0, 0.8], [0, 0.8], '-k', label='1:1')
ax2.plot([0, 4], [0, 4], '-k', label='1:1')

fig.tight_layout()
fig.savefig('pMELTS.png', dpi=200)
```
